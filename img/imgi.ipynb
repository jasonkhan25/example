```python
import math
class add():
    def __init__(self,name,x,y):
        self.name=name
        self.x=x
        self.y=y
add_1=add('ad1',9,1)
add_2=add('ad2',3,4)
dis=math.sqrt((add_1.x-add_2.x)**2+(add_1.y-add_2.y))
print(str(add_1.name),"= (",str((add_1.x)),",",str((add_1.y)),")")
print(str(add_2.name),"= (",str((add_2.x)),",",str((add_2.y)),")")
print(dis)
```
